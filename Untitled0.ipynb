```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
data.head()
```

перевести его в one hot

```python
data.loc[(data['whoAmI'] == 'human'),'Human'] = 1
```

```python
data.loc[(data['whoAmI'] == 'robot'),'robot'] = 1
```

```python
data['Human'] = data['Human'].fillna(0)
```

```python
data['robot'] = data['robot'].fillna(0)
```

```python
dt = data[['Human','robot']].copy()
data = data.drop('Human', axis =1)
data = data.drop('robot', axis =1)

dt.head()
```
